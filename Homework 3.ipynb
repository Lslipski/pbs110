##### (1) I flip a fair coin 12 times.Which is more likely: a) I get exactly 3 heads or b) that I get the third head on the fifth flip?

total ways to flip = 2^12 = 4096<br>

There are 12 flips and 3 of them need to come up heads<br>
$P(3 heads) = (12 choose 3) / 4096 = 220 / 4096 = 55 / 1024 = ~0.05$<br>

For this to happen, the first 4 flips need to have 2 heads and 2 tails. Then the 5th flip must be heads<br>
$P(fifth flip head) = ((4 choose 2)/ 2^4) * (1/2) = (6/16) * (1/2) = (3/8)*(1/2) = 3/16 = .1875$<br>

```r
# Alternatively, pascal method for 5th flip being the third heads
# p(N = n; r, p) =( (n-1)!/((r-1)!*(n-r)!)). * p^r * (1-p)^(n-r)
# n = 5; r= 3;
(factorial(4)/(factorial(2)*factorial(2))) * .5^3 * .5^2
```

So it is more likely to get the third flip heads.

##### (5) Generate a couple of sets of data for each of the binomial, the negative binomial, and the Poisson distribution varying parameters between the two for each distrbution. Include graphs from each of the three distributions plus the obtained means and sds and the predicted means and sds.

###### Binomial

```r
# Create the binomial distribution.
y <- rbinom(100, 50, 0.5)

# Create the binomial distribution.
q <- rbinom(100, 50, 0.25)

# Create the binomial distribution.
k <- rbinom(100, 50, 0.75)

par(mfrow=c(3,1))
# plot binomial
hist(y,main="rbinom(100, 50, 0.5)")
hist(q,main="rbinom(100, 50, 0.25)")
hist(k,main="rbinom(100, 50, 0.75)")
```

```r
#actual means: mean(n,p)
actual_mean_1 = mean(y)
actual_mean_2 = mean(q)
actual_mean_3 = mean(k)
#expected means
exp_mean_1 = 50*0.5
exp_mean_2 = 50*0.25
exp_mean_3 = 50*0.75

#actual stds
actual_std_1 = sd(y)
actual_std_2 = sd(q)
actual_std_3 = sd(k)
#expected stds
exp_std_1 = sqrt(50 * .5 * .5)
exp_std_2 = sqrt(50 * .25 * .75)
exp_std_3 = sqrt(50 * .75 * .25)

print(paste('Actual Mean Graph 1: ',actual_mean_1))
print(paste('Actual Mean Graph 2: ',actual_mean_2))
print(paste('Actual Mean Graph 3: ',actual_mean_3))

print(paste('Expected Mean Graph 1: ',exp_mean_1))
print(paste('Expected Mean Graph 2: ',exp_mean_2))
print(paste('Expected Mean Graph 3: ',exp_mean_3))

print(paste('Actual STD Graph 1: ',actual_std_1))
print(paste('Actual STD Graph 2: ',actual_std_2))
print(paste('Actual STD Graph 3: ',actual_std_3))

print(paste('Expected STD Graph 1: ',exp_std_1))
print(paste('Expected STD Graph 2: ',exp_std_2))
print(paste('Expected STD Graph 3: ',exp_std_3))
```

##### (7) This problem is equivalent to rolling a 32-sided die. It could be any of the 32 possible numbers on each roll. You keep rolling until you've rolled a 1,2,3,4,5,6,7, and an 8. You then note how many rolls it took. On average, how many trials should it take?

```r
reps = 10000
die = seq(1, 32, 1)
results = rep(0,reps)  

for (rep in 1:reps)  {     
  slots_to_fill = rep(0,8)  #need to fill all of these slots for the game to be over
  
  while (sum(slots_to_fill) < 8)  {  #keep going until all of the slots are filled i.e. game is over
    random_sample = sample(32, size=1, replace=TRUE)         #roll the die   
    if (random_sample <= 8) { #if the roll is 1-8
        slots_to_fill[random_sample] = 1 #fill the slot corresponding to the roll
    }
    results[rep] = results[rep] + 1      # increment the number of die rolls
  }
}
mean(results)  # take the mean number of rolls for all 10000 games
```
